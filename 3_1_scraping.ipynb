```python
import requests
import pandas as pd
```

`url: https://www.timesjobs.com`


```python
req = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=")
```

```python
req.status_code
```

```python
html = req.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html)
```

```python
souppre = soup.prettify()
```

```python
jobs_all = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
```

```python
titles = []
companies = []
descriptions = []
key_values = []
```

```python
for job in jobs_all:
    title = job.find('strong', class_ = 'blkclor').text
    company = job.find('h3', class_ = 'joblist-comp-name').text
    desc = job.find('ul', class_ = 'list-job-dtl clearfix').find('li').text
    key = job.find('span', class_ = 'srp-skills').text

    titles.append(title)
    companies.append(company)
    descriptions.append(desc)
    key_values.append(key)
```

```python
df = pd.DataFrame({'Titles': titles,
                   'Companies': companies,
                   'Descriptions': descriptions,
                   'Key Skills': key_values})
```

```python
df
```

```python
df.replace(to_replace='\r', value = '', regex = True, inplace = True)
df.replace(to_replace='\n', value = '', regex = True, inplace = True)
df.replace(to_replace='Job Description:', value = '', regex = True, inplace = True)
```

```python
df
```

```python
for i in range(1, 101):
    print(i)
    req = requests.get(f"https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=python&pDate=I&sequence=2&startPage={i}")
    html = req.text
    soup = BeautifulSoup(html)
    jobs_all = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
    
    titles = []
    companies = []
    descriptions = []
    key_values = []
    
    for job in jobs_all:
        title = job.find('strong', class_ = 'blkclor')
        company = job.find('h3', class_ = 'joblist-comp-name').text
        desc = job.find('ul', class_ = 'list-job-dtl clearfix').find('li').text
        key = job.find('span', class_ = 'srp-skills').text

        titles.append(title)
        companies.append(company)
        descriptions.append(desc)
        key_values.append(key)
        
    df1 = pd.DataFrame({'Titles': titles,
                   'Companies': companies,
                   'Descriptions': descriptions,
                   'Key Skills': key_values})
    
    df = pd.concat([df, df1])
```
